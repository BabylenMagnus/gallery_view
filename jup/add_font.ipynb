```python
a = np.array([[130.0, 325.0], [299.0, 325.0], [299.0, 448.0], [130.0, 448.0]])
a
```

```python
h = a[:, 1].max() - a[:, 1].min()
w = a[:, 0].max() - a[:, 0].min()

h, w
```

```python
image = Image.new('L', (int(h * 2), int(w * 2)), "white")
draw = ImageDraw.Draw(image)

message = "POP" 
myFont = ImageFont.truetype("/home/jjjj/Downloads/Blacknorthdemo-mLE25.otf", size=h)
_, _, w_, h_ = draw.textbbox((0, 0), message, font=myFont)
draw.text((0, 0), message, font=myFont, fill="black")
text_ = np.array(image)[:h_, :w_]
text_.shape
```

```python
text_ = Image.fromarray(text_).resize((int(h), int(w)), reducing_gap=3.0)
text_
```
